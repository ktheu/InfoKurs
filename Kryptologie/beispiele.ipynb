```python
def prepare(s):
    '''
    s: Klartext
    returns: String, Klartext ohne Umlaute, Satzzeichen, Leerzeichen, nur Großbuchstaben
    '''
    s = s.upper()
    s = s.replace('Ö','OE').replace('Ü','UE').replace('Ä','AE').replace('ß','SS')
    return ''.join([c for c in list(s) if c.isalpha()]) 
```
