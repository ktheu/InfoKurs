### Klassen - Beispiel: Quadratische Gleichung

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math

class QuadratischeGleichung:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        
    def __str__(self):
        return f'{self.a}x^2 {self.b:+}x {self.c:+} = 0'
    
    def diskriminante(self):
        return self.b*self.b - 4*self.a*self.c
    
    def anzahlLoesungen(self):
        if self.diskriminante() == 0:
            return 1
        elif self.diskriminante() < 0:
            return 0
        else:
            return 2
        
    def loesungen(self):
        if self.anzahlLoesungen() == 0: return []
        elif self.anzahlLoesungen() == 1:
            return [-self.b/(2*self.a)]
        else:
            sqd = math.sqrt(self.diskriminante())
            return [(-self.b + sqd)/(2*self.a), (-self.b - sqd)/(2*self.a)]
        
    def antwortSatz(self):
        res = self.loesungen()
        if len(res) == 0:
            print(f'Die Gleichung {self} hat keine Lösung.')
        elif len(res) == 1:
            print(f'Die Gleichung {self} hat die Lösung {res[0]:4.2f}.')
        else:
            print(f'Die Gleichung {self} hat die Lösungen {res[0]:4.2f} und {res[1]:4.2f}.')
        
 
```

```python
class Polynom2(QuadratischeGleichung):
    def __init__(self, a, b, c):
        super().__init__(a, b, c)
        
    def __str__(self):
        return f'f(x) = {self.a}x^2 {self.b:+}x {self.c:+}'
    
    def nullstellen(self):
        return self.loesungen()
    
    def val(self, x):
        return self.a*x*x + self.b*x + self.c
    
    def scheitelpunkt(self):
        return (-self.b/(2*self.a), self.c - (self.b*self.b)/(4*self.a))
    
    def plot(self, von=-10, bis=10):
        nst = self.nullstellen()
        fig, ax = plt.subplots()
        xs = np.linspace(von,bis,100)
    
        ax.plot(xs,self.val(xs),label=f'{self}');
        for x0 in nst:
            ax.scatter(x0,0,color='red')
            ax.text(x0+0.1,0.2,f'{x0:.2f}')
        sx, sy = self.scheitelpunkt()
        ax.scatter(sx, sy, color='blue')
        ax.text(sx+0.2,sy,f'({sx:.2f},{sy:.2f})')
        ax.axhline(y=0, color='k')
        ax.axvline(x=0, color='k')
        plt.grid(True)
        plt.legend()
        
        
```

```python
p = Polynom2(-3,12,-6)
print(p)
print(p.nullstellen())
print(p.scheitelpunkt())
p.plot(0,4)
```
